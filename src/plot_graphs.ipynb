```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(0)
import seaborn as sns
sns.set_style('darkgrid')
```

```python
def plot(list_mean_traces, list_std_traces, Xs, list_lbl, Xl, Yl, tit, fname='g.png', fsize=24):
    n_lines = len(list_lbl)
    plt.figure(figsize=(20,10))
        
    for i in range(n_lines):
        std = list_std_traces[i] 
        Ys = list_mean_traces[i]
        plt.plot(Xs, Ys, label=list_lbl[i])
        plt.fill_between(Xs, Ys-std, Ys+std, alpha=0.4)
    plt.title(tit, size=fsize)
    plt.xlabel(Xl, size=fsize)
    plt.ylabel(Yl, size=fsize)
    plt.legend(fontsize=fsize)
    plt.tight_layout()
    plt.savefig(fname)
```

```python
fname1 = '../out/trace_DQN_1.txt'
fname2 = '../out/trace_DQN_2.txt'
fname3 = '../out/trace_DQN_3.txt'
fname4 = '../out/trace_DQN_4.txt'
fname5 = '../out/trace_DQN_5.txt'
fname6 = '../out/trace_DQN_6.txt'
fname7 = '../out/trace_DQN_7.txt'
fname8 = '../out/trace_DQN_8.txt'
fname9 = '../out/trace_DQN_9.txt'
fname10 = '../out/trace_DQN_10.txt'

scores_trace = []
loss_trace = []
for fname in [fname1, fname2, fname3, fname4, fname5, fname6, fname7, fname8, fname9, fname10]:
    with open(fname, 'r') as f:
        lines_ = f.readlines()[2:]
        lines = [list(map(lambda s: s.strip(), l.split('|'))) for l in lines_]
        scores = [float(l[1].split(':')[1]) for l in lines]
        loss = [float(l[2].split(':')[1]) for l in lines]
        Xs = [float(l[0].split('episode')[0]) for l in lines]
    scores_trace.append(scores)
    loss_trace.append(loss)
n_g = len(scores_trace)
plt.figure(figsize=(20,10))
for i in range(n_g):
    plt.plot(Xs, scores_trace[i])
mean_scores1 = np.mean(scores_trace, axis=0)
std_scores1 = np.std(scores_trace, axis=0)
plt.title('Different traces of DQN', size=24)
plt.xlabel('Episodes', size=24)
plt.ylabel('Scores', size=24)
plt.tight_layout()
plt.savefig('DQN_traces.png')

mean_loss1 = np.mean(loss_trace, axis=0)
std_loss1 = np.std(loss_trace, axis=0)
# plt.figure(figsize=(20,10))
# for i in range(n_g):
#     plt.plot(Xs, loss_trace[i])
```
